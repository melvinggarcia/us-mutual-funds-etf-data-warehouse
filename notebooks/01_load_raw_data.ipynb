```python
import duckdb 
import pandas as pd 

# Connect to (or create) your local DuckDB database file
# This creates funds_warehouse.duckdb in your FundsDataWarehouse folder 
con = duckdb.connect('funds_warehouse.duckdb') 
 
# Drop table if it exists, then load ETF_prices.csv directly into DuckDB
con.execute("DROP TABLE IF EXISTS raw_etf_prices") 

con.execute(""" CREATE TABLE raw_etf_prices AS 
            SELECT * FROM read_csv_auto('data/raw/ETFs.csv') """)

# Verify row count
print(con.execute("SELECT count(*) as total_rows FROM raw_etf_prices").fetchdf())
```

```python
# Preview price data
print(con.execute("SELECT * FROM raw_etf_prices LIMIT 5").fetchdf())
```

```python
# Check the data range
print(con.execute(""" SELECT min(price_date) AS earliest_date, 
                  max(price_date) AS latest_date, count(DISTINCT fund_symbol) AS unique_funds 
                  FROM raw_etf_prices """).fetchdf())
```

```python
# Check column types
print(con.execute("DESCRIBE raw_etf_prices").fetchdf())
```

```python
# Load ETF profile data
con.execute("DROP TABLE IF EXISTS raw_etf_info") 

con.execute(""" CREATE TABLE raw_etf_info AS 
            SELECT * FROM read_csv_auto('data/raw/ETFs.csv') """)

print(con.execute("SELECT count(*) as total_etfs FROM raw_etf_info").fetchdf())
print(con.execute("SELECT * FROM raw_etf_info LIMIT 3").fetchdf())
```

```python
# Check data completeness
print(con.execute(""" SELECT count(*) AS total,
                  count(fund_symbol) AS has_symbol, 
                  count(fund_category) AS has_category, 
                  count(fund_family) AS has_family, 
                  count(inception_date) AS has_inception_date, 
                  count(fund_return_ytd) AS has_ytd_return, 
                  count(fund_sharpe_ratio_5years) AS has_sharpe_5yr 
                  FROM raw_etf_info""").fetchdf())
```

```python
# Close connection
con.close()
```
